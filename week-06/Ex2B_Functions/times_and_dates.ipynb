# Lab 2: Working with DateTime Functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now() # Creating a variable that represents current date and time
```

```python
# strftime() means string format time
# Using .strftime() to format the date into a readable style.
# %A means full weekday name, %B means full month name, %d means day number, and %Y is for 4-digit year
print("Today is", today.strftime("%A, %B %d, %Y"))
```

# My Birthday

```python
birthday = datetime.date(2004, 8, 26) # Storing my birthday as a variable
```

```python
# %m is for the month, %d is for the day, %Y is for the 4-digit year
print("My birthday is", birthday.strftime("%m/%d/%y"))
```

# Next 90 Days

```python
# days=90 finds the date 90 days from current date. Future date is stored as a variable
ninety_d = today + datetime.timedelta(days=90)
print("90 days from today is", ninety_d.strftime("%B %d, %Y"))
# strftime is for formattting, timedelta is for calculating
```

# Time for Dinner

```python
dinner_time = datetime.time(20, 30) # Defining variable for dinner
# Python uses military time, 20 is the hour and 30 is the minutes
```

```python
# %I formats it to 12-hour, %M for minutes and %p for AM or PM
print("Let's meet for dinner at", dinner_time.strftime("%I:%M %p"))
```

## Math with Integers and Floats

### Understanding the difference between `/` and `//` (and how this impacts datatypes)

Write an expression to illustrate the answer to each question below. Make sure to run the cell to check your results.

```python
# What is the result of 2 times 2?
# The result of 2 * 2 is 4 because multiplication between two integers returns an integer.

2 * 2
```

```python
# What is the result of 4 divided by 2, using regular division?
# The result of 4 / 2 is 2.0 because regular division returns a float always regardless if
# theres a remainder or not.

4/2
```

```python
# What is the result of 4 divided by 2, using floor division?
# The result of 4 / / 2 is 2 because it divides evenly and returns an integer. If there 
# was a remainder, it would drop the remainder and give the quotient as an integer.

4 // 2
```

```python
# Ok, how about 2 times 2.5?
# The result is 5.0 because multiplication between an integer and float returns a float,
# even if numerically there's no value after the decimal.

2 * 2.5
```

```python
# What's 5 divided by 2 using regular division?
# The result is 2.5 because regular division returns a float.

5 / 2
```

```python
# What's 5 divided by 2 using floor division?
# The result is 2 because floor division drops the remainder and keeps only the smallest
# nearest full integer. It won't round up.

5 // 2
```

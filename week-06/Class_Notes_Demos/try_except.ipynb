## What Are Exceptions?

In Python, **exceptions** are runtime errors that disrupt normal program execution. Instead of crashing, Python allows us to **handle errors gracefully** using exception handling.

- When an error occurs, the interpreter stops and generates an error message
- Errors raised by the interpreter are known as **exceptions** — special objects used to manage errors that arise while a program is running
- Anticipating exceptions and handling them makes your programs more **robust** and provides a better user experience

> **Tip:** An exception should only ever be an *exception to a rule* — if you can use other programming logic to prevent it, that is the better approach!

### Common Exception Types

| Exception | Cause |
|---|---|
| `ZeroDivisionError` | Dividing by zero |
| `ValueError` | Wrong value type (e.g. entering text instead of a number) |
| `TypeError` | Operation on incompatible types |
| `FileNotFoundError` | File doesn't exist |
| `IndexError` | List index out of range |
| `NameError` | Variable not defined |
| `KeyError` | Dictionary key not found |
| `SyntaxError` | Invalid Python syntax |

---

## Order of Flow — The 4 Blocks
try:     → code that may/may not produce errors
except:  → when error arises, this block executes
else:    → when NO error arises, this block executes
finally: → ALWAYS runs, whether error occurs or not


### Best Practices

1. Use the `try` block to wrap code that might raise an exception.
2. Follow with one or more `except` blocks — specify the exception type (e.g., `ZeroDivisionError`, `ValueError`).
3. Optionally add an `else` block — runs only if **no** exceptions occurred.
4. Use a `finally` block for cleanup code (e.g., closing files) — runs regardless.
5. Avoid a bare `except` block unless absolutely necessary — it can mask unexpected errors.
6. Use the `raise` keyword to explicitly trigger exceptions with a custom message.

## Example 2: Try, Except, Clse, and Finally together

This is the **full structure** of exception handling in Python.

````python
# Example of try, except, else, and finally in Python

try:
    # Ask the user for a number
    number = int(input("Enter a number: "))

    # Divide 100 by the number
    result = 100 / number

except ValueError:
    # Runs if the user enters non-numeric data
    print("Error: Please enter a valid number.")

except ZeroDivisionError:
    # Runs if the user enters 0
    print("Error: Cannot divide by zero.")

else:
    # Runs only if no errors occur
    print(f"The result is: {result}")

finally:
    # Always runs whether an error occurs or not
    print("Program execution completed.")
````

## Example 3: Error Handling with User Input

````python
# Error handling — validate user age input
try:
    age = int(input("Enter your age: "))
    print("Next year you will be", age + 1)
except ValueError:
    print("That is not a valid age.")
````

## Example 4: Try-Except Inside a Loop (Re-Prompting)

The previous examples catch the exception but give no chance to try again. Place `try-except` inside a **loop** to keep re-prompting until valid input is received.

````python
# Keep asking until a valid positive number is entered
while True:
    try:
        number = int(input("Enter a positive number: "))
        if number > 0:
            break
        else:
            print("Number must be positive.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

print("Valid number received:", number)
````

## Deciding Which Errors to Report

These are **predictable situations** where you should use `try-except`:

| Situation | Example |
|---|---|
| Invalid user input | User types text instead of a number |
| Working with files | File may not exist |
| Risky calculations | Dividing by a user-provided number |
| List/dictionary access | Uncertain keys or indexes |
| External systems | Databases, APIs |

### Example: Handling a Missing File

````python
# Error handling — file that may not exist
try:
    file = open("data.txt", "r")
    content = file.read()
except FileNotFoundError:
    print("File not found. Please check the filename.")
# Without the except, it will give a traceback error — this just makes the error more readable
````

## Key Takeaways

- **Error handling** helps code flow smoothly with no interruptions or crashes
- **Without try-except**, risky code will produce a traceback that crashes the program
- **Multiple `except` blocks** can handle different errors separately
- **`raise`** manually triggers an exception inside a function with a custom message
- **In data analysis**, use error handling for `ValueError` (non-numeric column values) or `KeyError` (missing columns in a dataset)

### ❌ Without try-except (This will cause an error!)

```python
a = 12
b = 6
result = a / b
print(result)                               # prints 2.0
print("I have reached the end of the line") # this is what we are excepting
```
> Leaving the code like this **will cause an error** because `try` and `except` wasn't used.

---

### ✅ Fixed with Order of Flow

```python
try:
    a = 12
    b = 6
    result = a / b
    print(result)
except:
    print("We caught an error")
else:
    print("Hurray, we don't have any errors")
finally:
    print("I have reached the end of the line")
```

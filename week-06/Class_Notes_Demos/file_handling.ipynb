File handling in Python allows you to perform operations such as creating, reading, writing, appending and deleting files. It is essential for tasks like storing data permanently and processing files efficiently.

> **Proper sequence is always: Open → Read/Write → Close**
---

## Key File Operations

| Operation | Code | What it does |
|---|---|---|
| Open | `f = open('data.txt', 'r+')` | Opens the file and creates a file object. `'r+'` opens for reading AND writing |
| Read | `content = f.read()` | Reads all content from the file. Cursor moves to the end |
| Read one line | `f.readline()` | Reads just one line at a time — call it multiple times for multiple lines |
| Read all lines | `f.readlines()` | Returns a **list** of strings, each string is one line |
| Write | `f.write('Hello Python!\n')` | Writes (or **overwrites**) content to the file |
| Append | `f.write('This is appended.\n')` | Adds content to the **end** of existing content |
| Close | `f.close()` | Closes the file to free up memory |
| With statement | `with open() as f:` | Opens AND automatically closes the file when done — **best practice!** |
| Check exists | `os.path.exists()` | Checks if a file exists before opening or deleting |
| Delete | `os.remove()` | Deletes the file — needs `import os` first |

---

## Python File Modes

| Mode | Name | What it does |
|---|---|---|
| `'r'` | Read | Opens file for reading. **Default mode.** Returns error if file doesn't exist |
| `'w'` | Write | Creates or **overwrites** file. Creates the file if it doesn't exist |
| `'a'` | Append | Adds to end of existing file. Creates the file if it doesn't exist |
| `'x'` | Create | Creates a new file. Returns **error** if file already exists |
| `'b'` | Binary | Used for binary files (images, etc.) — combine with other modes e.g. `'wb'` |

> **Quick rule:** `'w'` wipes the file and starts fresh. `'a'` keeps what's there and adds to it. If you're not sure which to use, ask yourself — do I want to keep the old content?
---

## Example 1: Writing to a File

Using `with open()` is the preferred way — it automatically closes the file when the block finishes so you don't have to remember to call `f.close()`.

```python
# Writing to a file
# 'w' mode creates the file if it doesn't exist, or overwrites it if it does
with open("students.txt", "w") as file:
    file.write("Alice\n")
    file.write("Bob\n")

# with open closes the file automatically when you finish adding what you needed
```

## Example 2: Reading from a File

```python
# Reading from a file
# 'r' mode is the default — opens for reading only
with open("students.txt", "r") as file:
    content = file.read()
    print(content)

# Output:
# Alice
# Bob
```

## Example 4: Creating a New File and Writing to it

```python
# Open file for writing — creates example.txt if it doesn't exist
# WARNING: if example.txt already exists, 'w' will wipe it and start fresh!
with open("example.txt", "w") as file:
    file.write("Hello, Python!")
```

## Example 5: Appending to a File

Append adds to the **end** of what's already there — it does NOT overwrite.

```python
# Append text to a file — adds to end without erasing existing content
with open("example.txt", "a") as file:
    file.write("\nNew Line added.")
```

## 🗑️ Deleting Files

Python has no built-in delete function — you need to `import os` first.

| Function | What it does |
|---|---|
| `os.remove("file.txt")` | Deletes a file. Raises `FileNotFoundError` if it doesn't exist |
| `os.rmdir("folder")` | Deletes an empty directory. Raises error if not empty |
| `os.rename("old.txt", "new.txt")` | Renames a file |
| `os.replace("old.txt", "new.txt")` | Renames a file — silently replaces if destination already exists |

> 💡 **Always check if the file exists before deleting** to avoid errors — use `os.path.exists()` wrapped in a `try-except`.

```python
import os

# Best practice: check if it exists before trying to delete
if os.path.exists("example.txt"):
    os.remove("example.txt")
    print("File deleted.")
else:
    print("File not found — nothing to delete.")
```

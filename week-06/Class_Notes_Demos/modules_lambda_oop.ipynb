## Random Module 
The `random` module is a built-in Python module that generates random numbers and performs random operations.

````python
import random 
# 1. random.randint(a, b) - Random integer between a and b(inclusive)
print("randint(1, 10):", random.randint(1, 10))
````

````python
#2. random.random() - Random float between 0.0 and 1.0
print("random():", random.random())
````

````python
# 3. random.choice(seq) - Random single element from a sequence
fruits = ["apple", "banana", "cherry"]
print("choice(fruits):", random.choice(fruits))
````

````python
# 4. random.chocies(seq, k=n) - List of random k elements (with replacement)
print("choices(fruits, k=3):", random.choices(fruits, k=3))
# k gives us how manu results we want from the list
````

````python
# 5. random.sample(seq, k=n) - List of k unique random elements (without replacement)
print("sample(fruits, k=2):", random.sample(fruits, k=2))
````

````python
# 6. random.shuffle(seq) - Shiffles a list in place
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print("shuffled numbers:", numbers)
# shuffle does not work on tuples because they are immutable (unchangeable)
````

## DateTime Module

The `datetime` module is a built-in Python module used to work with dates and times.

### Data Types in the datetime module:
| Data Type | Stores | Syntax |
|---|---|---|
| `date` | year, month, day | `date(year, month, day)` |
| `time` | hour, minute, second, microsecond | `time(hour, minute, second)` |
| `datetime` | date + time together | `datetime(year, month, day, hour, minute)` |
| `timedelta` | difference between dates | `timedelta(days=7)` |
| `timezone` | timezone offset | `timezone(timedelta(hours=-5))` |

````python
from datetime import date, time, datetime, timedelta, timezone
````

````python
# date(year, month, day)
today = date(2026, 5, 9)
print("Date:", today)
````

````python
# time(hour, minute, second, microsecond)
current_time =  time(14, 30, 45)
print("Time:", current_time)
````

````python
# datetime(year, month, day, hour, minute, second)
current_datetime = datetime(2026, 5, 9, 14, 30)
print("Datetime:", current_datetime)
````

````python
# timezone(timedelta(hours=offset))
est = timezone(timedelta(hours=-5))
dt_est = datetime(2026, 5, 9, 14, 30, tzinfo=est)
print("Timezone:", dt_est)
````

## Statistics Module

The `statistics` module is a built-in Python module used to calculate mathematical statistics from numerical data.

````python
import statistics
````

````python
# Sample
data = [10, 20, 20, 40, 50]
print("=== Python Statistics Module Demo ===\n")

# mean() - average value
print("mean:", statistics.mean(data))

# median() - middle value
print("median:", statistics.median(data))

# mode() - most common value
print("mode:", statistics.mode(data))

# stdev() - standard deviation
print("stdev:", statistics.stdev(data))

# variance() - variance
print("variance:", statistics.variance(data))
````

## Lambda Function

**Definition:** A small anonymous one-line function that can take any number of arguments but contains only one expression.

### Syntax:
`lambda parameters: expression`

### Key Rules:
- Can have **any number of parameters**
- Can only have **one expression**
- No need for `def` or `return`
- Stored in a variable to be reused

### Compared to a Regular Function:
```python
# Regular function
def doubler(n):
    return n * 2

# Same thing as a lambda
doubler = lambda n: n * 2
```
### When to use lambda:
- Simple one-line operations
- When you don't want to write a full function
- Common in data analytics for quick calculations

````python
# lambda function to double a number
doubler = lambda n: n * 2
print(doubler(5))
# 5 is being passed in the function call
````

````python
# lambda function to add a number
add_numbers = lambda a, b: a + b
print(add_numbers(3, 7))
````

````python
# lambda function to determine largest number
largest = lambda x, y: x if x > y else y
print(largest(12, 8))
````

### Real World Example — Monthly Loan Payment

**Problem:** A person takes out a $25,000 loan at 0.5% monthly interest 
over 60 months (5 years). Calculate the monthly payment.

**Formula:**
M = P * (r * (1 + r)^n) / ((1 + r)^n - 1)

Where:
- M = Monthly payment (what we are solving for)
- P = Principal loan amount = $25,000
- r = Monthly interest rate = 0.005 (0.5%)
- n = Number of payments = 60 (5 years)

````python
# WHY lambda? Because this is a simple one-line formula that doesn't need
# a full def function — lambda keeps it clean and concise
# WHY 3 parameters? Because the formula needs 3 inputs: loan amount, rate, and time
monthly_payment = lambda P, r, n: P * (r * (1 + r) ** n) / ((1 + r) ** n - 1)
#                        ↑  ↑  ↑                                   
#                        P  r  n  ← parameters passed in below

# Example values
principal = 25000       # P = Loan amount in dollars
monthly_rate = 0.005    # r = 0.5% monthly interest rate
payments = 60           # n = 60 monthly payments
                        # 5 years x 12 months = 60 total payments

# WHY do we store it in payment variable?
# So we can reuse the result in the print statement below
# WHY pass principal, monthly_rate, payments?
# These match P, r, n in the lambda — Python maps them in order
payment = monthly_payment(principal, monthly_rate, payments)

print(f"Monthly Payment: ${payment:.2f}")
````

## List Comprehension

**Definition:** A concise way to create lists using a single line of code instead of writing a full loop.

### Syntax:
`[expression for variable in iterable]`

### Key Parts:
| Part | Meaning |
|---|---|
| `[ ]` | Creates a list |
| `expression` | What to do with each item |
| `for variable in iterable` | The loop that goes through each item |

````python
# Long way - regular loop
numbers = [1, 2, 3, 4, 5]
squares = []                      # empty list to store results
for n in numbers:                 # loop through each number
    squares.append(n ** 2)        # square each number and .append adds them to list
    print(squares)

# Short way - list comprehension
# Does the exact same thing in one line!
squares = [n ** 2 for n in numbers]
#           ↑         ↑
#        expression   loop
print(squares)        # [1, 4, 9, 16, 25]
````

## OOP — Object Oriented Programming

**Definition:** OOP combines a group of variables (attributes) and functions (methods) into a unit called an "object." These objects are organized into classes where individual objects can be grouped together.

A **class** in Python is a blueprint or template used to create objects. It defines:
- Attributes(variables): data the object stores
- Methods (Functions): Actions the object can perform

### Structure of a Class:

    class ClassName:
    
        def __init__(self, attribute1, attribute2):  # constructor, always first
            self.attribute1 = attribute1             # store attribute 1
            self.attribute2 = attribute2             # store attribute 2

        def method_name(self):                  # method, runs when called
            # do something with the attributes

    # Then OUTSIDE the class:

    # Step 1 - Create object
    object1 = ClassName(value1, value2)

    # Step 2 - Call method
    object1.method_name()

### Key Terms:
| Term | Meaning | Example |
|---|---|---|
| `class` | Blueprint/template | `class Student:` |
| `object` | The actual thing made from the class | `student1 = Student("Alice", 95)` |
| `__init__` | Constructor, sets up the object | `def __init__(self, name, grade)` |
| `self` | Refers to the current object | `self.name = name` |
| `attribute` | Data the object stores | `self.name, self.grade` |
| `method` | Function inside the class | `def display_info(self)` |

### Demo for Class

````python
# Define a Python class
# You can have any number of methods but you must have a constructor

class Student: # creates a class named Student

    # Constructor method - runs automatically when object is created
    # WHY __init__? So every student object starts with a name and grade ready to use
    # self refers to the current object being created
    # WHY self? So Python knows which specific student we are talking about
    # NOTE: self is always the first parameter of every method
    # Python automatically passes it in behind the scenes when you create an object
    def __init__(self, name, grade):
    #             ↑
    #        self is here so Python knows WHICH object is being created

        self.name = name      # creates an attribute called name
        self.grade = grade    # creates an attribute called grade
        # NOTE: self. on these lines means "attach this data to THIS specific object"
        # WHY self.name and self.grade? So each student
        # stores their OWN name and grade separately
        # student1.name = "Alice", student2.name = "Brian" — kept separate!

    # Method to display student information
    # WHY a method? So we can reuse this display logic for any student
    # only runs when we call it
    # NOTE: self is also required here so the method knows
    # which student's data to display
    def display_info(self):
        print(f"Student Name: {self.name}")   # prints THIS student's name
        print(f"Grade: {self.grade}")          # prints THIS student's grade
        print("---------------------------")

# Create 2 student objects from the Student class
# WHY? Each object is a separate student with their own data
# each student gets their own name and grade
student1 = Student("Alice", 95)  # Alice → name, 95 → grade
student2 = Student("Brian", 88)  # Brian → name, 88 → grade

# Display the student objects using f strings
# we access each student's attributes using dot notation
# WHY dot notation? student1.name means "get name FROM student1"
print(f"Student 1: {student1.name} {student1.grade}")
print(f"Student 2: {student2.name} {student2.grade}")
print()

# Call the method for each student
# WHY .display_info()? To run the method we defined inside the class
student1.display_info()
student2.display_info()
````

### Create your own Class
This class stores and displays information about makeup products.


````python
# Define a Makeup class
# You can have any number of methods but you must have a constructor

class Makeup: # creates a class named Makeup

    # Constructor method - runs automatically when object is created
    # WHY __init__? So every makeup object starts with its details ready to use
    # self refers to the current object being created
    # WHY self? So Python knows which specific product we are talking about
    def __init__(self, brand, product, shade, price):
        self.brand = brand      # creates an attribute called brand
        self.product = product  # creates an attribute called product
        self.shade = shade      # creates an attribute called shade
        self.price = price      # creates an attribute called price
                                # WHY self.brand etc? So each product
                                # stores its OWN details separately

    # Method to display makeup information
    # WHY a method? So we can reuse this display logic for any makeup product
    # only runs when we call it
    def display_info(self):
        print(f"Brand:   {self.brand}")
        print(f"Product: {self.product}")
        print(f"Shade:   {self.shade}")
        print(f"Price:   ${self.price:.2f}")
        print("---------------------------")

# Create 3 makeup objects from the Makeup class
# WHY? Each object is a separate product with its own details
makeup1 = Makeup("MAC", "Lipstick", "Ruby Woo", 21.00)
makeup2 = Makeup("Fenty Beauty", "Foundation", "130N", 38.00)
makeup3 = Makeup("Charlotte Tilbury", "Blush", "Pillow Talk", 45.00)

# Display the makeup objects using f strings
# WHY dot notation? makeup1.brand means "get brand FROM makeup1"
print(f"Product 1: {makeup1.brand} - {makeup1.product} in {makeup1.shade}")
print(f"Product 2: {makeup2.brand} - {makeup2.product} in {makeup2.shade}")
print(f"Product 3: {makeup3.brand} - {makeup3.product} in {makeup3.shade}")
print()

# Call the method for each product
# WHY .display_info()? To run the method we defined inside the class
makeup1.display_info()
makeup2.display_info()
makeup3.display_info()
````

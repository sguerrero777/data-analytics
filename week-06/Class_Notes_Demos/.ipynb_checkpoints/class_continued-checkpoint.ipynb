# Classes Continued — Encapsulation & Inheritance

---

## 📌 Quick Recap — How to Create a Class

- To create a class, use keyword `class`, capitalize the name, and end with a colon
- You need at least one special method: `__init__`
- `__init__` initializes (sets up) the object — it runs automatically when you create a new object
- `self` always comes first in every method — it refers to the current object you're working with, never change it

## Default Class Attribute

A **default class attribute** is a variable inside a class (within `__init__`) that is automatically assigned a pre-set value for every instance created — without requiring the user to pass that value in as an argument.

| Type | Belongs To | Shared? |
|---|---|---|
| Class attribute | The CLASS itself | ✅ Yes, all objects share it |
| Instance attribute | ONE specific object | ❌ No, unique to each object |

## Encapsulation

**Encapsulation** = bundling data (attributes) and methods together inside a class, and controlling who can access or change that data.

Two key ideas:
- **Private attributes** → use `__` (double underscore) to HIDE the data — only accessible inside the class
- **Getters** → a method that lets you safely READ a private attribute
- **Setters** → a method that lets you safely UPDATE a private attribute with validation before changing

> **Why bother**? Because you don't want someone accidentally setting a grade to 500 or a salary to -1. The setter acts like a bouncer checking before letting anything in.

```python
# demonstrates encapsulation
class Student:
    school = "YearUp United"   # Class attribute — shared by ALL students

    def __init__(self, name, grade, track):
        self.name = name             # Public attribute — anyone can access
        self.__grade = grade         # PRIVATE attribute (encapsulated!)
        self.track = track           # Public attribute

    # GETTER - controlled way to READ private grade
    # you can't do student1.__grade directly from outside the class
    def get_grade(self):
        return self.__grade

    # SETTER - controlled way to UPDATE private grade (with validation)
    # notice how we check BEFORE actually changing the value
    def set_grade(self, new_grade):
        if 0 <= new_grade <= 100:
            self.__grade = new_grade
        else:
            print("❌ Invalid grade! Must be between 0 and 100.")

    def display_info(self):
        print(f"School: {Student.school}")
        print(f"Student Name: {self.name}")
        print(f"Grade: {self.__grade}")   # accessing private attr inside class is fine
        print(f"Track: {self.track}")
        print("--------------------")

# Create student objects
student1 = Student("Alice", 95, "Software Development")
student2 = Student("Brian", 88, "Data Analytics")

# ❌ This would FAIL - grade is private, can't access from outside!
# print(student1.__grade)

# Use the getter instead
print(student1.get_grade())     # Output: 95

# Use the setter to update safely
student1.set_grade(98)          # Works! Valid grade
student1.set_grade(150)         # ❌ Invalid! Must be between 0 and 100

print()
student1.display_info()
student2.display_info()
```

## Inheritance

**Inheritance** = when a child class gets all the methods and attributes of a parent class automatically.

> Think of it like genetics — a child inherits traits from a parent but can also have their own unique traits on top of that.

- `super()` → lets the child class call the parent's methods without naming the parent directly

| Type | Description |
|---|---|
| Single Inheritance | Child inherits from ONE parent class |
| Multiple Inheritance | Child inherits from MULTIPLE parent classes |
| Multilevel Inheritance | Child inherits from a parent that already inherited from another parent (grandparent chain) |

```python
print()

# --- PARENT CLASS ---
# This is the base blueprint that ALL student types share
class Student:

    # Class attributes (shared by ALL students no matter the track)
    school  = "YearUp Academy"
    diploma = "High School Diploma"

    # Constructor - instance attributes unique to each student
    def __init__(self, name, age, gpa):
        self.name = name
        self.age  = age
        self.gpa  = gpa

    # Shared method - inherited by every child class automatically
    def display_info(self):
        print(f"  School  : {Student.school}")
        print(f"  Diploma : {Student.diploma}")
        print(f"  Name    : {self.name}")
        print(f"  Age     : {self.age}")
        print(f"  GPA     : {self.gpa}")
```

```python
# --- CHILD CLASS 1 - App Dev ---
# AppDev inherits from Student — gets everything Student has
# PLUS adds its own track-specific attributes (language, project)
class AppDev(Student):                         # inherits from Student

    track = "Application Development"          # class attribute (track-level)

    def __init__(self, name, age, gpa, language, project):
        super().__init__(name, age, gpa)       # calls Student.__init__ so we
                                               # don't have to repeat name/age/gpa
        self.language = language               # unique to AppDev students
        self.project  = project                # unique to AppDev students

    def display_info(self):                    # overrides parent display_info
        super().display_info()                 # runs Student.display_info first
        print(f"  Track   : {AppDev.track}")
        print(f"  Language: {self.language}")
        print(f"  Project : {self.project}")
```

```python
# --- CHILD CLASS 2 - Data Analytics ---
# Same idea as AppDev but with different extra attributes
class DataAnalytics(Student):                  # also inherits from Student

    track = "Data Analytics"                   # class attribute

    def __init__(self, name, age, gpa, tool, dataset):
        super().__init__(name, age, gpa)       # inherit name, age, gpa from Student
        self.tool    = tool                    # unique to DataAnalytics students
        self.dataset = dataset                 # unique to DataAnalytics students

    def display_info(self):
        super().display_info()                 # print the Student info first
        print(f"  Track   : {DataAnalytics.track}")
        print(f"  Tool    : {self.tool}")
        print(f"  Dataset : {self.dataset}")
```

```python
# --- CREATE STUDENT OBJECTS ---
student1 = AppDev(
    name="Alice",
    age=20,
    gpa=3.8,
    language="Python",
    project="Student Portal App"
)

student2 = AppDev(
    name="Marcus",
    age=21,
    gpa=3.5,
    language="JavaScript",
    project="Job Board Website"
)

student3 = DataAnalytics(
    name="Brian",
    age=22,
    gpa=3.7,
    tool="Pandas",
    dataset="NYC Sales Data"
)

student4 = DataAnalytics(
    name="Sofia",
    age=20,
    gpa=3.9,
    tool="Matplotlib",
    dataset="US Census Report"
)
```

```python
# --- DISPLAY ALL STUDENT REPORTS ---
# putting all students in a list so we can loop through them
# instead of calling display_info() one by one
all_students = [student1, student2, student3, student4]

for student in all_students:
    print("=" * 40)
    student.display_info()

print("=" * 40)

# --- ACCESS CLASS ATTRIBUTES DIRECTLY ---
# these belong to the class itself, not any one student
print()
print(f"School    : {Student.school}")
print(f"Diploma   : {Student.diploma}")
print(f"Track 1   : {AppDev.track}")
print(f"Track 2   : {DataAnalytics.track}")

# --- CHECK INHERITANCE ---
# isinstance() checks if an object belongs to a class
# useful for figuring out what type of object you're working with
print()
print(f"student1 is a Student?        {isinstance(student1, Student)}")
print(f"student3 is a Student?        {isinstance(student3, Student)}")
print(f"student1 is an AppDev?        {isinstance(student1, AppDev)}")
print(f"student1 is a DataAnalytics?  {isinstance(student1, DataAnalytics)}")
```

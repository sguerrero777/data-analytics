## Weekly Sales Statistics Analyzer

```python
import statistics
```

```python
name = input("Analyst name: ")
region = input("Region: ")
```

```python
sales = []
for day in range(1, 8):
    figure = input(f"Enter sales for day {day}:")
    sales.append(float(figure))
```

```python
def analyze_sales(name, region, sales):
    mean = statistics.mean(sales)
    median = statistics.median(sales)
    mode = statistics.mode(sales)
    stdev = statistics.stdev(sales)
    total = sum(sales)
    high = max(sales)
    low = min(sales)
    return mean, median, mode, stdev, total, high, low
analysis = analyze_sales(name, region, sales)
    
```

```python
print(f"""
=== Sales Analysis Report ===
Analyst: {name}
Region:  {region}

Mean:    ${analysis[0]:,.2f}
Median:  ${analysis[1]:,.2f}
Mode:    ${analysis[2]:,.2f}
Stdev:   ${analysis[3]:,.2f}
Total:   ${analysis[4]:,.2f}
High:    ${analysis[5]:,.2f}
Low:     ${analysis[6]:,.2f}

Raw Data: {sales}
""")
```

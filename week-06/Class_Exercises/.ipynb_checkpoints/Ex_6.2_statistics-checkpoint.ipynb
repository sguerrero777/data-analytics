## Statistics Module Quick Reference

| Function | Returns | What it does |
|---|---|---|
| `statistics.mean(data)` | float | Arithmetic average — same as `sum(data)/len(data)` |
| `statistics.median(data)` | float | Middle value when sorted — not affected by outliers |
| `statistics.mode(data)` | value | Most common value — raises `StatisticsError` if no unique mode |
| `statistics.stdev(data)` | float | **Sample** std deviation (divides by n-1) — use when data is a sample |
| `statistics.pstdev(data)` | float | **Population** std deviation (divides by n) — use when data is the full population |
| `statistics.variance(data)` | float | Sample variance — variance is just stdev squared |
| `sum(data)` | float | Total — built-in, no import needed |
| `max(data)` | float | Highest value — built-in |
| `min(data)` | float | Lowest value — built-in |

> **stdev vs pstdev:** Use `stdev()` when your data is a **sample** (a slice of a bigger dataset). Use `pstdev()` when you have the **full population** (all the data that exists). For weekly sales of 7 days, `stdev()` is correct because those 7 days are a sample of all possible sales days.

---

## Step 1 — Import the Statistics Module

```python
import statistics
# statistics is part of Python's standard library — no pip install needed
# it MUST be imported at the very top before anything else
# calling statistics.mean() without importing first causes a NameError
```

## Step 2 — Collect User Input

```python
# Collect analyst name and region — input() always returns a STRING
name = input("Analyst name: ")
region = input("Region: ")
```

## Step 3 — Collect 7 Days of Sales Data

```python
# Start with an empty list to store each day's sales
sales = []

for day in range(1, 8):   # range(1, 8) gives us 1, 2, 3, 4, 5, 6, 7
    figure = input(f"Enter sales for day {day}:")
    sales.append(float(figure))  # float() converts the string input to a decimal number
                                  # WITHOUT float(), statistics functions would throw a TypeError
                                  # because they can't do math on strings

# Shortcut — the same thing in ONE line using list comprehension:
# sales = [float(input(f" Day {i+1}: $")) for i in range(7)]
# range(7) gives 0-6, so we do i+1 to display Day 1 through Day 7
```

## Step 4 — Define the analyze_sales Function

```python
def analyze_sales(name, region, sales):
    # statistics module functions — need the import at the top
    mean   = statistics.mean(sales)    # average of all 7 days
    median = statistics.median(sales)  # middle value when sorted
    mode   = statistics.mode(sales)    # most common value
    stdev  = statistics.stdev(sales)   # how spread out the numbers are

    # built-in functions — no import needed
    total  = sum(sales)   # add all values together
    high   = max(sales)   # highest single day
    low    = min(sales)   # lowest single day

    # returning multiple values — Python automatically packs these into a TUPLE
    # a tuple is like a list but cannot be changed after creation
    # on the calling line you can unpack them: mean, median, mode... = analyze_sales(...)
    return mean, median, mode, stdev, total, high, low

# Call the function and unpack ALL 7 returned values at once
# the order here must match the order in the return statement above!
analysis = analyze_sales(name, region, sales)

# You could also unpack like this instead of using analysis[0], analysis[1] etc:
# mean, median, mode, stdev, total, high, low = analyze_sales(name, region, sales)
```

## Step 5 — Print the Formatted Report

```python
# Triple quotes """ let us write a multi-line string without \n everywhere
# f before the quotes makes it an f-string so we can use {variables} inside
# :.2f formats the number as a float with exactly 2 decimal places
#   ex: 1296.607 becomes 1296.61
#   the : starts the format, . means decimal, 2 means 2 places, f means float

print(f"""
=== Sales Analysis Report ===
Analyst: {name}
Region:  {region}

Mean:    ${analysis[0]:,.2f}
Median:  ${analysis[1]:,.2f}
Mode:    ${analysis[2]:,.2f}
Stdev:   ${analysis[3]:,.2f}
Total:   ${analysis[4]:,.2f}
High:    ${analysis[5]:,.2f}
Low:     ${analysis[6]:,.2f}

Raw Data: {sales}
""")

# Notice analysis[0] = mean, analysis[1] = median, etc.
# The index matches the order values were returned in analyze_sales()
# The , in :,.2f adds comma separators for thousands ex: $1,296.61
```

## Mode Warning

`statistics.mode()` raises a `StatisticsError` if no single value appears more than once (called **multimodal data**). Here's how to handle it safely:

```python
# Safe version of mode using try-except
# if every sales number is different, mode will raise StatisticsError
# this catches that and prints a friendly message instead of crashing

try:
    mode = statistics.mode(sales)
    print(f"Mode: ${mode:,.2f}")
except statistics.StatisticsError:
    print("Mode: No unique mode — all values appear the same number of times")
```

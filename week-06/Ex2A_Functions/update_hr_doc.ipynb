## Lab 3. Employee Data 
This lab exercise makes updates to information extracted from 
an employee database. 

```python
hr_list = [ 
('0123', 'HR', 'Rebecca Yang', '69000'),  
('0121', 'IT', 'Mark Blick', '67000'),  
('0068', 'IT', 'Kahna Larsen', '112000'),  
('0234', 'OPS', 'Jim Smith', '54000') 
]
```

```python
# Mark reporting surname change to Blick-Hawley

hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
# hr_list grabs Mark's info at index 1 and rewrites the entire tuple with his updated last name
# since tuples are immutable, I have to rewrite the entire item with Mark's new information
# changing just one portion of a tuple is not allowed, only for lists. The entire tuple is replaced
```

```python
print(hr_list)
```

```python
# Jims role change

hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
print(hr_list)
```

```python
# Displaying Contents Using format()

for i in hr_list:
    print(f"{i[0]} | {i[1]} | {i[2]} | {float(i[3]):,.2f}")

# Loop goes through each item in the list and print function uses f-string to format them a specific way using index position
```

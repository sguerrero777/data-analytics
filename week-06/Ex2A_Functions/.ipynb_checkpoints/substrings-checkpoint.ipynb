# Lab 1. Working with Names - Parsing Strings

```python
test_name = 'Sharleen Guerrero' # variable for my name
```

```python
def parse_and_display (name):   # def creates a function. name is a placeholder that will be filled in with whatever variable you hand to the function when you actually run it.
    for i in range(len(name)):  # for loop that uses len() to count characters and range() to loop through each index position in the name
        if name [i] == ' ': # checks if the character at position i is a space
            first = name[:i] # everything before the space becomes the first name
            last = name[i+1:] # everything after the space becomes the last name
            print(f"Full name: {name}") # prints the full name
            print(f"First name: {first}") # prints the first name
            print(f"Last name: {last}") # prints the last name
```

```python
parse_and_display(test_name)
```

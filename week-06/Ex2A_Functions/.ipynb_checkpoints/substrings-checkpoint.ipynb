# Lab 1. Working with Names - Parsing Strings

```python
test_name = 'John Doe' # variable for name
```

```python
# Parsing a string means reading through it and pulling out specific pieces of information.
# Code does this by finding spaces between the words.
def parse_and_display (name): # def creates a function. name is a parameter that only 
#exists inside the function and gets its value when the function is called.
# for i in range() loops through however many characters len counted in the string, i is a
# variable that keeps track of index position, can be anything
    for i in range(len(name)): 
        if name [i] == ' ': # checks where the character at position i is a space
            first = name[:i] # everything before the space becomes the first name
            last = name[i+1:] # everything after the space becomes the last name
            print(f"Full name: {name}") # prints the full name
            print(f"First name: {first}") # prints the first name
            print(f"Last name: {last}") # prints the last name
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
parse_and_display(test_list[0])
```

```python
parse_and_display(test_list[1])
```

```python
parse_and_display(test_list[2])
```

```python
parse_and_display('Sharleen Guerrero')
```

## Bonus:
Create a modified version of your function called parse_and_display2() that will work 
with names made of one, two or three words. Output should include middle name if applicable.

```python
def parse_and_display2(name):
    if name.count(' ') == 0: # .count counts how many spaces are in the string
        first = name # if there's no spaces, it makes sense the name is equal to itself
        print(f"First name: {first}")
    elif name.count(' ') == 1:
        for i in range(len(name)): # len counts the characters, range creates the positions, and the loop visits each one
            if name[i] == ' ': # che cks the position of the index where there's a space
                first = name[:i] # : says anything before the index position of the space
                last = name[i+1:] # adds 1 after the position of the space to exclude it and mvoe on to next index
        print(f"First name: {first}")
        print(f"Last name: {last}")
    elif name.count(' ') == 2: 
        spaces = [] # a space is an character with an index position, and i can only one index postition, a list is needed to store multiple
        for i in range(len(name)):
            if name[i] == ' ': # loop through every character until a space is found
                spaces.append(i) # append ads a new item at the end of the list, in this case, it saves the index position of each space
            
        #slicing happens AFTER the loop, when both space positions are saved in the list
        first = name[:spaces[0]] # : is from the beginning, and spaces[0] is wherever the index position of the first sapce is. This grabs everything before the first space, so the first name
        middle = name[spaces[0]+1:spaces[1]] # +1 is going to the next index postion to exlude the position of space[0], the first space, and then grabbing everything until the second space begins, where [1] is the position of the second space, but in the list, it is the second index
        last = name[spaces[1]+1:] # space[1]+1 starts at the next index position after the second space, and the colon at the end slices everything after that space
       
        print(f"First name: {first}")
        print(f"Middle name: {middle}")
        print(f"Last name: {last}")
```

```python
parse_and_display2('Lorde')
```

```python
parse_and_display2('Billie Eilish')
```

```python
parse_and_display2('Megan Thee Stallion')
```

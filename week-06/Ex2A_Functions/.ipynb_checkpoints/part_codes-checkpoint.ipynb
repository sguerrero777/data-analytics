# Exercise 2.A Using Built-in Functions 
## Lab 2. Parsing Strings – Part Codes

```python
acme_part = 'ACME:123-L'
di_part = 'DI:12345-M'
ace_part = 'ACE:1-12'
```

```python
def get_supplier_code(full_string): # Defining a function to get the supplier code from the full string
    colon = full_string.find(":")   # finds and stores the index position of wherever the colon is
    supplier = full_string[:colon]  # grabs every character before the position of the colon
    return supplier                 # returns that part of the string
```

```python
print(get_supplier_code(acme_part))
print(get_supplier_code(di_part))
print(get_supplier_code(ace_part))
```

```python
def get_product_num(full_string):              # Defining a function to get the product code, which is between the colon and hyphen
    colon = full_string.find(":")              # finds and stores the index position of wherever the colon is 
    hyphen = full_string.find("-")             # finds and stores the index position of wherever the hypen is
    product_code = full_string[colon+1:hyphen] # grabs everything after the colon position and stops at the hyphen position
    return product_code                        # returns that part of the string
```

```python
print(get_product_num(acme_part))
print(get_product_num(di_part))
print(get_product_num(ace_part))
```

```python
def get_size(full_string):          # Defining a function to get size of the product
    hyphen = full_string.find("-")  # finds and stores the index position of where the hyphen is
    size = full_string[hyphen+1:]   # grabs everything after the hyphen, +1 excludes hyphen
    return size                     # returns that part of the string
```

```python
print(get_size(acme_part))
print(get_size(di_part))
print(get_size(ace_part))
```

```python
print(f"For {acme_part}, the supplier is {get_supplier_code(acme_part)}.")
print(f"Part number {get_product_num(acme_part)} is available in size {get_size(acme_part)}.")
```

```python
print(f"For {di_part}, the supplier is {get_supplier_code(di_part)}.")
print(f"Part number {get_product_num(di_part)} is available in size {get_size(di_part)}.")
```

```python
print(f"For {ace_part}, the supplier is {get_supplier_code(ace_part)}.")
print(f"Part number {get_product_num(ace_part)} is available in size {get_size(ace_part)}.")
```

**Bonus**: Try creating a variable that contains all three part codes as a list. 

```python
part_codes = ["ACME:123-L", "DI:12345-M", "ACE:1-12"]

for i in part_codes:
    print(f"For {i}, the supplier is {get_supplier_code(i)}.")
    print(f"Part number {get_product_num(i)} is available in size {get_size(i)}.")
```

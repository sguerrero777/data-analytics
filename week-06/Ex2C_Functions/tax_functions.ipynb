## Lab 3. Defining Tax Functions

This lab demonstrates the creation and calling of custom Python functions to automate payroll tax calculations. Skill used are:

- **Defining Functions**: Creating reusable blocks of code with parameters and return values.
- **Conditional Logic**: Using `if/elif/else` statements to apply different tax rates based on withholding codes.
- **Mathematical Operations**: Implementing specific tax formulas for withholding adjustments.
- **Formatting**: Using f-strings to display financial data

```python
def get_soc_sec_tax(gross):  # function with a parameter for gross pay
    return(gross * .062)     # adds a tax rate of 6.2% on gross pay
```

```python
def get_medicare_tax(gross):  # function with a parameter of gross pay
    return(gross * .0145)     # adds a tax rate of 1.45% on gross pay
```

```python
def get_federal_tax(gross, withhold_code): # function with two parameters, gross pay and withholding code
    if withhold_code == 0:
        tax_rate = .23
    elif withhold_code == 1:
        tax_rate = .21
    elif withhold_code  == 2:
        tax_rate = .195
    elif withhold_code == 3:
        tax_rate = .185
    else:
        tax_rate = .18 - ((withhold_code - 4) * .005)
    return gross * tax_rate
```

```python
# Testing the functions (Person1: gross pay $750 withholding code 0)
print("--- (Person1: gross pay $750 withholding code 0) ---")
print(
    f"The Social Security tax     : ${get_soc_sec_tax(750):.2f}\n"
    f"The Medicare tax            : ${get_medicare_tax(750):.2f}\n"
    f"The federal tax withholding : ${get_federal_tax(750, 0):.2f}\n"
)
```

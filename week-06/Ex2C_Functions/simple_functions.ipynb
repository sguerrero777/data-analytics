# Exercise 2.C Creating and Calling Functions
## Lab 1. Defining Custom Functions

```python
def favorite_things(name, movie, music):   # Creating function with three parameters
    return (f"Name: {name}\nFavorite movie: {movie}\nFavorite music/band: {music}")
    # returning 3 lines using \n to show the output of parameters

# calling the function
print(favorite_things("Sharleen", "Michael", "Michael Jackson"))
```

```python
def why_im_here(purpose): # Creating function with purpose as the parameter
    return purpose        # Returning my purpose

# calling the function
print(why_im_here("I'm here to gain scalable knowledge and skills in data analytics and earn a work-based experience and get my foot in the tech sector."))
```

```python
def favorite_place(place, why): # Creating function with two parameters
    return(f"One of my favorite places is {place} because {why}.")
```

```python
# calling the function
print(favorite_place("Kin Ramen", "it has the best ramen and matcha lava cake in New York"))
```

```python
# calling the function
print(favorite_place("Atlantic City", "it has the funnest resorts in New Jersey"))
```

### Exercise 2.E Using Functions for Analyzing Data 
This notebook analyzes a small sales dataset using Python functions. 
Each function takes the full dataset as its first argument. 

```python
sales_data = [
    ('Laptop',     'East',  1200.00),
    ('Phone',      'West',   850.00),
    ('Tablet',     'East',   400.00),
    ('Monitor',    'South',  650.00),
    ('Keyboard',   'West',    75.00),
    ('Laptop',     'South', 1150.00),
    ('Phone',      'East',   900.00),
    ('Tablet',     'West',   425.00),
    ('Monitor',    'East',   680.00),
    ('Keyboard',   'South',   80.00),
    ('Laptop',     'West',  1300.00),
    ('Phone',      'West',  820.00),
]
```

## Lab 1. Total Sales by Region

```python
def region_total(sales_data, region): # Defining a function to return the total of all sales amounts for a specific region
    return sum(amount for _, r, amount in sales_data if r == region)
    # A generator expression that loops and filters inside the function call
    # _ means for any value thats innecessary, in this case ignores products
    # it knows to ignore products because it's first in the generator, matching it's position in the tuple
    # calculates for sales by region, not total sales in dataset
```

```python
print(region_total(sales_data, 'East')) 
print(region_total(sales_data, 'West')) 
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of Sales by Region

```python
def region_count(sales_data, region): # Defining a function to return the number of records for each region
    return sum(1 for _, r, _ in sales_data if r == region)
    # _ ignores product and amount because only r is needed to check if the region matches, 
    # then count 1 every time r == region is true
```

```python
print(region_count(sales_data, 'East')) 
print(region_count(sales_data, 'West')) 
print(region_count(sales_data, 'South'))
```

## Lab 3. Average Sale Amount by Region

```python
def region_average(sales_data, region): # Defining a function to return the average sale amount by region
    avg = 0 # setting average to 0 to return if a region has no records and avoid division by 0 error
    if region_count(sales_data, region) > 0: # Sales count for the region must be greater than 0
        avg = region_total(sales_data, region) / region_count(sales_data, region) # average is calculated by dividing total sales per region and count of sales per region
    return avg
```

```python
print(f"{region_average(sales_data, 'East'):.2f}")
print(f"{region_average(sales_data, 'West'):.2f}")
print(f"{region_average(sales_data, 'South'):.2f}")
print(region_average(sales_data, 'North'))
```

## Lab 4. Filter by Sale Amount

```python
def sales_above(sales_data, threshold):
    # filters through sales_data, i represents each full tuple
    # returns a new list containing only tuples where
    # the amount (index position [2]) is greater than the threshold
    return list(i for i in sales_data if i[2] > threshold)
```

```python
for record in sales_above(sales_data, 500):
    print(record)
```

```python
# Bonus
def sales_below(sales_data, threshold):
    return list(i for i in sales_data if i[2] < threshold)
```

```python
for record in sales_below(sales_data, 500):
    print(record)
```

## Lab 5. Top and Bottom Sale

```python
def top_sale(sales_data): # Defining a function to return the single record(tuple) with highest sale amount
    return max(sales_data, key=lambda i: i[2]) 
    # Uses the max function with key as an argument where it returns the tuple where the index position of 
    # the sale amount is the highest

def bottom_sale(sales_data):
    return min(sales_data, key=lambda i: i[2])

print(top_sale(sales_data))
print(bottom_sale(sales_data))
```

- The **top sale** was a `Laptop` in the West region for **$1300**

- The **bottom sale** was a `Keyboard` in the West region for **$75**

## Lab 6. Apply a Discount

```python
def apply_discount(sales_data, discount_pct):
    # map() applies the lambda to every tuple in the list
    # since tuples are immutable, we can't change i[2] directly
    # so we build a brand new tuple with all three values
    # i[0] and i[1] stay the same (product and region)
    # i[2] is recalculated with the discount applied
    return list(map(lambda i: (i[0], i[1], i[2] * (1 - discount_pct)), sales_data))
```

```python
discounted_data = apply_discount(sales_data, 0.15)

print("--- Before discount ---")
print(f"{sales_data[0]}\n{sales_data[1]}\n{sales_data[2]}\n")
print("--- After 15% discount ---")
print(f"{discounted_data[0]}\n{discounted_data[1]}\n{discounted_data[2]}\n")
```

## Lab 7. Region Summary

```python
def unique_regions(sales_data): # Defining a function that displays all region names no duplicates
    return sorted(set(r for _, r, _ in sales_data))
    # grabs the region, ignores the product and amount in dataset
    # sorted() sorts it in alphabetical order
    # set() removes duplicates
```

```python
print(f"Unique Regions: {unique_regions(sales_data)}")
```

This function could be useful for determining how many active regions a large dataset spans without manually having to search each row.

## Lab 8. Region Summary

```python
def region_summary(sales_data, region):
    region_data = [i for i in sales_data if i[1] == region]
    # for every tuple (i) in the list, only keep it if index[1] (region) matche

    total = region_total(region_data, region)   # calculate total for that region
    count = region_count(region_data, region)   # calculate count for that region
    avg = region_average(region_data, region)   # calculate average for that region
    top = top_sale(region_data)                 # find top sale for that region

    print(f"=== {region} Region ===")
    print(f"Records       : {count}")
    print(f"Total Sales   : ${total:,.2f}")
    print(f"Average Sale  : ${avg:,.2f}")
    print(f"Top Sale      : {top[0]} (${top[2]:,.2f})")
    # top is the tuple returned by top_sale()
    # top[0] grabs index position 0 (product name) from the tuple
    # top[2] grabs index position 2 (sale amount) from the tuple and formats it with .2f
```

```python
(region_summary(sales_data, 'East'))
# no print needed because it is already inside the function. If the function uses return, 
# wrap it in print when calling as rule of thumb
```

```python
for region in unique_regions(sales_data):  # loops through unique regions (no duplicates)
    region_summary(sales_data, region)      # calls all functions and displays results for each region
```
